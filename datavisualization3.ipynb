```

import numpy as mp
import matplotlib.pyplot as plt
import pandas as pd
```

```
df = pd.read_csv('http://13.234.66.67/summer19/datasets/bank.csv')
df.head(10)
```

```
plt.bar(df.Age,df.Balance,color="blue")
plt.xlabel("Age")
plt.ylabel("Balance")
```

```
plt.stackplot(df.Age,df.EstimatedSalary)
plt.xlabel("Age")
plt.ylabel("EstimatedSalary")
```

```
plt.scatter(df.Age,df.EstimatedSalary)
plt.xlabel("Age")
plt.ylabel("EstimatedSalary")
plt.grid()
```

```
df.groupby('Geography').size().plot('pie',shadow=True, label='Geography',explode=[0.04,0,0] ,autopct='%d%%')
```

```

plt.bar(df.Gender,df.Balance)
plt.xlabel("Gender")
plt.ylabel("Balance")
```

```
df.groupby('Gender').size().plot('pie',shadow=True,label='Gender',explode=[0,0.05],autopct='%d%%')
```

```
plt.fill(df.Gender,df.CreditScore)
plt.xlabel("Gender")
plt.ylabel("CreditScore")
plt.grid()
```
